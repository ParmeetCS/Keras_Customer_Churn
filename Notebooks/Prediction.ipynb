# Customer Churn Prediction - Inference

This notebook demonstrates how to load the trained model and the unified preprocessor to make predictions on new customer data.

```python
import pandas as pd
import numpy as np
import pickle
from tensorflow.keras.models import load_model
```

## 1. Load Model and Preprocessor

```python
model = load_model('model.h5')

with open('preprocessor.pkl', 'rb') as file:
    preprocessor = pickle.load(file)

print(f'Expected feature names: {preprocessor.feature_names_in_}')
```

## 2. Define Input Data
*We provide raw data exactly as it was formatted in the CSV file.*

```python
input_data = {
    'Gender': 'Male',
    'Senior Citizen':'Yes',
    'Partner': 'Yes',
    'Dependents': 'Yes',
    'Tenure Months': 1,
    'Phone Service': 'Yes',
    'Multiple Lines': 'Yes',
    'Internet Service': 'Fiber optic',
    'Online Security': 'Yes',
    'Online Backup': 'Yes',
    'Device Protection': 'Yes',
    'Tech Support': 'Yes',
    'Streaming TV': 'Yes',
    'Streaming Movies': 'Yes',
    'Contract': 'Month-to-month',
    'Paperless Billing': 'Yes',
    'Payment Method': 'Electronic check',
    'Monthly Charges': 110.0,
    'Total Charges': 110.0,
    'CLTV': 100,
}

# Convert to DataFrame (crucial step!)
input_df = pd.DataFrame([input_data])
print('Input DataFrame preview:')
```

## 3. Preprocess the Input Data

```python
input_scaled = preprocessor.transform(input_df)
print(f'Processed Input shape: {input_scaled.shape}')
print(f'Processed Input array preview:\n{input_scaled}')
```

## 4. Predict Churn

```python
prediction_prob = model.predict(input_scaled)[0][0]
prediction_class = int(prediction_prob >= 0.5)

print(f'\nModel Prediction Probability: {prediction_prob:.4f}')
print(f'Model Prediction Class: {prediction_class}') 

if prediction_class == 1:
    print('ALERT: The customer is likely to churn (High Risk).')
else:
    print('SAFE: The customer is likely to remain with the company.')
```
